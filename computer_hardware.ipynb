```python
df.drop(['vendor_name','model_name'],axis=1,inplace=True)
df
```

```python
ip=df.drop('ERP',axis=1)
op=df.ERP
```

```python
ip
```

```python
op.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2,random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
yp=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
b=mean_squared_error(y_test,yp)
c=r2_score(y_test,yp)
print('Mean_squared_error:',b)
print('r2_score:',c)
```
